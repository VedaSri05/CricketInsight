### Import required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the cleaned dataset

```python
df = pd.read_csv("../data/bcci_odi_stats_cleaned.csv")
```

```python
df.head()
```

### Summary of numerical columns

```python
print("\nSummary Statistics:")
df.describe()
```

### Top 5 players with the highest total runs

```python
print("\nTop 5 Players by Total Runs:")
df.nlargest(5, "Total Runs")[["Name", "Total Runs"]]
```

### Top 5 players with the highest batting average

```python
print("\nTop 5 Players by Batting Average:")
df.nlargest(5, "Average Score")[["Name", "Average Score"]]
```

### Top 5 players with the best strike rate

```python
print("\nTop 5 Players by Strike Rate:")
df.nlargest(5, "Strike Rate")[["Name", "Strike Rate"]]
```

### Players with most centuries and fifties

```python
print("\nTop 5 Players by Centuries:")
df.nlargest(5, "Hundreds")[["Name", "Hundreds"]]
```

```python
print("\nTop 5 Players by Fifties:")
df.nlargest(5, "Fifties")[["Name", "Fifties"]]
```

### Visualization - Top 10 Players by Total Runs

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Total Runs", y="Name", data=df.nlargest(10, "Total Runs"), palette="viridis")
plt.xlabel("Total Runs")
plt.ylabel("Player Name")
plt.title("Top 10 Players by Total Runs")
plt.show()
```

### Visualization - Top 10 Players by Batting Average

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Average Score", y="Name", data=df.nlargest(10, "Average Score"), palette="coolwarm")
plt.xlabel("Batting Average")
plt.ylabel("Player Name")
plt.title("Top 10 Players by Batting Average")
plt.show()
```

### Visualization - Top 10 Players by Strike Rate

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Strike Rate", y="Name", data=df.nlargest(10, "Strike Rate"), palette="Blues")
plt.xlabel("Strike Rate")
plt.ylabel("Player Name")
plt.title("Top 10 Players by Strike Rate")
plt.show()
```

### Correlation Heatmap

```python
# Exclude non-numeric columns
numeric_df = df.select_dtypes(include=["number"])

# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

### Distribution Plot of Total Runs

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Total Runs"], bins=20, kde=True, color="purple")
plt.xlabel("Total Runs")
plt.ylabel("Frequency")
plt.title("Distribution of Total Runs")
plt.show()
```

### Comparing Players' Boundary-Hitting Ability (Fours vs Sixes)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Fours"], y=df["Sixes"], hue=df["Name"], palette="coolwarm", edgecolor='k')
plt.xlabel("Fours")
plt.ylabel("Sixes")
plt.title("Comparison of Fours and Sixes Hit by Players")
plt.legend([],[], frameon=False)  # Hide legend for clarity
plt.show()
```

### Most Efficient Players (Strike Rate vs Average Score)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Strike Rate"], y=df["Average Score"], hue=df["Name"], palette="viridis", edgecolor='k')
plt.xlabel("Strike Rate")
plt.ylabel("Batting Average")
plt.title("Most Efficient Players (Strike Rate vs Batting Average)")
plt.legend([],[], frameon=False)
plt.show()
```
